```python
import pandas as pd
```

```python
df = pd.read_excel('movies.xls', sheet_name=0, index_col=0)
df.head()
```

```python
df1 = pd.read_excel('movies.xls', sheet_name=1, index_col=0)
df2 = pd.read_excel('movies.xls', sheet_name=2, index_col=0)
# reading sheet 1 and 2
```
